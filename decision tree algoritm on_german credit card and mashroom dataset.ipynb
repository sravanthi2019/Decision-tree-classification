## Splitting the data

I splitted the data 90% training and 10% for testing. so that the target variable is balanced in both.

## Model

C50 package is used for Decision tree algorithm. which used Entropy and information gain for splitting the tree.



```r
install.packages("C50")
```

```r
# Training the model
library(C50)
credit_model <- C5.0(credit_train[,-c(17)], factor(credit_train$default))
summary(credit_model)
```

From the above, The size of decision tree is 42.It used 16 attributes out of 21 attributes. for tree building.Checking balance is the root node, it means the information gain from this variable is maximal compared to all.

```r
# predicting for test dataset.
credit_pred <- predict(credit_model, credit_test)
```

```r
install.packages('gmodels')
library(gmodels)
CrossTable(factor(credit_test$default), credit_pred,
prop.chisq = FALSE, prop.c = FALSE, prop.r = FALSE,
dnn = c('actual default', 'predicted default'))
```

From above cross table,out of 65 non default 55 predicted as non default and 10 predicted wrong. Out of 35 default 13 predicted as  default. The model is not very accurate as it gave just 68% accuracy.

```r
install.packages('lubridate')
library(lubridate)
library(caret)
confusionMatrix(factor(credit_test$default),factor(credit_pred))
```

The accuarcy of the model is 68%. the specificity is just 56%. we canimprove the model performence by boosting.

## Bossting

As the above model performence is less, we can improve performance by boosting.boosting is rooted in the notion that by combining a number of weak performing learners, we can create a team that is much stronger than any of the learners alone.

For boosting we need to add an additional trials parameter indicating the number of separate decision trees to use in the boosted team. The trials parameter sets an upper limit; the algorithm will stop adding trees if it recognizes that additional trials do not seem to be improving the accuracy.

```r
# by adding trails
credit_boost10 <- C5.0(credit_train[-17], factor(credit_train$default),
trials = 10)
```

```r
credit_boost10
```

```r
summary(credit_boost10)
```

```r
credit_boost_pred10 <- predict(credit_boost10, credit_test)
```

```r
CrossTable(credit_test$default, credit_boost_pred10,
prop.chisq = FALSE, prop.c = FALSE, prop.r = FALSE,
dnn = c('actual default', 'predicted default'))
```

with boosting total numbee of non default preicted increased to 59 and default predicted increased to 18.eventually the performence got increased.

```r
confusionMatrix(factor(credit_test$default),factor(credit_boost_pred10))
```

The  accuracy increased to 77%. The specificty is increased to 75%.

## Checking model with penalizing.

The business has given false negative has a cost of 4 versus a false positive's cost of 1. SO penalized those values with error cost.

```r
matrix_dimensions <- list(c("1", "2"), c("1", "2"))
names(matrix_dimensions) <- c("predicted", "actual")
```

```r
matrix_dimensions
```

```r
error_cost <- matrix(c(0, 1, 4, 0), nrow = 2,
dimnames = matrix_dimensions)
```

```r
error_cost
```

```r
credit_cost <- C5.0(credit_train[,-c(17)], factor(credit_train$default),
costs = error_cost,na.action = na.pass)
```

```r
summary(credit_cost)
```

```r
credit_cost_pred <- predict(credit_cost, credit_test)
```

```r
CrossTable(credit_test$default, credit_cost_pred,
prop.chisq = FALSE, prop.c = FALSE, prop.r = FALSE,
dnn = c('actual default', 'predicted default'))
```

## Decision Rule Making on Mashroom Dataset

## Data Extraction

Data extracted from Machine learning with R Github.It has 8124 records with 23 variables. The target variable is type whether it is e(edible) or p(poisonous) of mashroom with given features like shape, surface, color etc. all of them are factor variables.

```r
mushrooms <- read.csv("C:\\Users\\kgajj\\AppData\\Local\\Temp\\Temp1_Machine-Learning-with-R-datasets-master.zip\\Machine-Learning-with-R-datasets-master\\mushrooms.csv", stringsAsFactors = TRUE)
```

```r
str(mushrooms)
```

 The variable Veil type has same level for all the types. so it does not make any affect on the model. so i deleted it.

```r
# deleting one column
mushrooms$veil_type <- NULL
```

```r
# checking the class variables.how many are  e type or p type.
table(mushrooms$type)
```

More than 50% of the mashrooms are edible and remaining are poisonous.

Since ZeroR ignores all of the features and simply predicts the target's mode.its rule would state that all the mushrooms are edible. Obviously,this is not a very helpful classifier, because it would leave a mushroom gatherer sick or dead. We will use the 1R implementation in the RWeka package called OneR().



```r
install.packages("RWeka")
library(RWeka)
mushroom_1R <- OneR(type ~ ., data = mushrooms)
```

```r
mushroom_1R
```

```r
summary(mushroom_1R)
```

Considering that the learner utilized only a single feature, it did reasonably well.if one avoids unappetizing smells when foraging for mushrooms, they will almost avoid a trip to the hospital

## Improving Model Performence

the process of training a JRip() model is very similar to how we previously trained a OneR() model. This is one of the pleasant benefits of the functions in the RWeka package; the syntax is consistent across algorithms, which makes the process of comparing a number of different models very simple.

```r
mushroom_JRip <- JRip(type ~ ., data = mushrooms)
```

```r
mushroom_JRip
```

```r
summary(mushroom_JRip)
```

The JRip() classifier learned a total of nine rules from the mushroom data. An easy way to read these rules is to think of them as a list of if-else statements, similar to programming logic. The first three rules could be expressed as:
 If the odor is foul, then the mushroom type is poisonous
 If the gill size is narrow and the gill color is buff, then the mushroom type is poisonous
 If the gill size is narrow and the odor is pungent, then the mushroom type is poisonous.
 Else, the mushroom is edible

## Summary

The two classification methods that use so-called "greedy" algorithms to partition the data according to feature values. Decision trees use a divide and conquer strategy to create flowchart-like structures, while rule learners separate and conquer data to identify logical if-else rules. Both methods produce models that can be interpreted without a statistical background.

### Credit dataset:

The C5.0 algorithm to create a tree to predict whether a loan applicant will default. Using options for boosting and cost-sensitive errors, we were able to improve our accuracy and avoid risky loans that would cost the bank more money.

### Mashroom Datset:

The two rule learners, 1R and RIPPER, to develop rules to identify poisonous mushrooms. The 1R algorithm used a single feature to achieve 99 percent accuracy in identifying potentially fatal mushroom samples. On the other hand, the set of nine rules generated by the more sophisticated RIPPER algorithm correctly identified the edibility of each mushroom.
